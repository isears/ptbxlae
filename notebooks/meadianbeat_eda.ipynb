```python
import neurokit2 as nk
from ptbxlae.dataprocessing.ptbxlDS import PtbxlDS
```

```python
SAMPLE_IDX = 45
ds = PtbxlDS(root_folder='../data/', lowres=False)
sig = ds[SAMPLE_IDX]
from ptbxlae.dataprocessing import plot_raw_data
plot_raw_data(sig, {'fs': 500})
```

```python
#nk.ecg_plot(sig, sampling_rate=500)
cleaned = nk.ecg_clean(sig[7], sampling_rate=500)
cleaned.shape
```

```python
rpeaks = nk.ecg_findpeaks(cleaned, sampling_rate=500)
rpeaks
```

```python
segments = nk.ecg_segment(cleaned, rpeaks=rpeaks['ECG_R_Peaks'], sampling_rate=500, show=True)
```
