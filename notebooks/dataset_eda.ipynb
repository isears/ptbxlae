```python
from ptbxlae.dataprocessing.ptbxlDS import ptbxlDS
from ptbxlae.dataprocessing import plot_raw_data
import torch
```

```python
ds = ptbxlDS(root_folder="../data/ptbxl", lowres=False)
len(ds)
```

```python
plot_raw_data(ds[100]['waveform'], {'fs': 500})
```

```python
DL_DEMO_LIMIT = 5

dl = torch.utils.data.DataLoader(
    ds,
    num_workers=2,
    batch_size=4,
)

for batchnum, sigs in enumerate(dl):
    print(batchnum, sigs['waveform'].shape)

    if batchnum > DL_DEMO_LIMIT:
        break
```
